```python
with open("result.txt", "r" , encoding="ascii") as f:
    lst = f.readlines()

lst = [int(line.strip()) for line in lst]

# 40
cnt_40 = 0
for i in lst:
    if i == 40:
        cnt_40 += 1

from matplotlib import pyplot as plt
import matplotlib as mpl
import numpy as np
lst = np.array(lst)
print(lst.std(ddof=True), lst.min(), lst.max(), lst.mean(), lst.mean() / 80, cnt_40 / len(lst))
plt.figure(figsize=(10,10) )
mpl.rcParams["font.size"] = 19
plt.hist(lst, bins=30,edgecolor='#FFFFFF', density=True)
plt.xlabel("Count of changed bits")
plt.ylabel("Frequency distribution")

plt.show()
```
